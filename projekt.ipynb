# $\text{1. Wprowadzenie}$ 

Piłka nożna, najpopularniejszy sport na świecie, jest częścią naszego życia od setek lat, generując miliony, a nawet miliardy dolarów zysku rocznie. W centrum tego globalnego fenomenu leży rzut karny – moment pełen napięcia i emocji, który często decyduje o wyniku meczu. Piłkarze i trenerzy od zawsze zastanawiali się, jak uderzyć piłkę, aby zwiększyć szanse na strzelenie gola.

W tym projekcie podejmiemy się analizy uproszczonego modelu strzału z rzutu karnego, próbując odpowiedzieć na pytanie, jakie techniki wykonania strzału – takie jak kąt kopnięcia i prędkość piłki – wpływają na trajektorię i ostateczne trafienie do bramki. Wykorzystamy równania różniczkowe do modelowania ruchu piłki, biorąc pod uwagę siły, takie jak grawitacja i opór powietrza, a następnie zastosujemy metody numeryczne, w tym metodę Eulera i metodę Rungego-Kutty czwartego rzędu, do symulacji trajektorii piłki.

Nasz model będzie przedstawiony w dwóch wymiarach (2D) i dla uproszczenia przeprowadzimy symulację bez udziału bramkarza. Przez połączenie naukowego podejścia do symulacji z pasją do piłki nożnej, nasz projekt ma na celu nie tylko poszerzenie rozumienia fizyki ruchu w sporcie, ale także dostarczenie praktycznych wskazówek, które mogą pomóc piłkarzom w doskonaleniu ich techniki strzałów z rzutów karnych.


![rysunek poglądowy](rysunek.png)

Z naszego modelu wynika, że piłka po kopnięciu będzie poruszać się w górę i w prawo, więc możemy to uznać jako ruch w osiach $X$ oraz $Y$. <br> W ruchu po osi $X$, piłce "przeszkadzać będzie"  opór powietrza, w osi $Y$ działać na piłkę bedzie opór powietrza i grawitacja. Opór powietrza jest wprost proporcjonalny do prędkości piłki. <br> Stąd otrzymujemy dwa równania:

\begin{align*} m\frac{\operatorname{d}^2\!x}{\operatorname{d}\!t^2} &= -k\left(\frac{\operatorname{d}\!x}{\operatorname{d}\!t}\right)\\  m\frac{\operatorname{d}^2\!y}{\operatorname{d}\!t^2} &= -mg -k\left(\frac{\operatorname{d}\!y}{\operatorname{d}\!t}\right)  \\
\end{align*}
gdzie:
-  $m$ to masa piłki. 
- $k$ to współczynnik oporu powietrza. 
- $g$ to przyspieszenie grawitacyjne 
- $x$ to położenie piłki w czasie na osi $X$
- $y$ to położenie piłki w czasie na osi $Y$


Dzieląc obustronnie przez masę dostajemy równania:
\begin{align*}
\frac{\operatorname{d}^2\!x}{\operatorname{d}\!t^2} &= -\frac{k}{m}\left(\frac{\operatorname{d}\!x}{\operatorname{d}\!t}\right)\\  \frac{\operatorname{d}^2\!y}{\operatorname{d}\!t^2} &= -g -\frac{k}{m}\left(\frac{\operatorname{d}\!y}{\operatorname{d}\!t}\right) \\
\end{align*}

# $\text{2. Rozwiązanie analityczne}$ 

Dla ułatwienia zapisu przyjmiemy, że  $\frac{k}{m} = \beta$ 

Zaczniemy od rozwiązania równania dla osi $X$. Wiemy, że przyspieszenie to pochodna prędkości, więc na początek wyznaczymy równanie dla  $v_{x}$  

\begin{align*} \frac{\operatorname{d}\!v_{x}}{\operatorname{d}\!t} &= -\beta \cdot v_{x} \\

    \frac{\operatorname{d}\!v_{x}}{v_{x}} &= -\beta \operatorname{d}\!t \\
    \int\frac{\operatorname{d}\!v_{x}}{v_{x}}  &= -\int \beta \operatorname{d}\!t \\
    \ln(v_{x}) &= -\beta \cdot t + \ln(C) \\
    v_{x} &= C \cdot \mathrm{e}^{-\beta t}
\end{align*}    

Wiemy, że piłka w momencie kopnięcia ma nadaną pewną prędkość początkową $v_{0}$. 
Z rozkładu tej prędkości na składowe wynika, że w kierunku osi $X$, piłka została kopnięta z prędkością początkową $v_{x_0}= v_{0}\cos{\alpha}$.

Podstawiając ten warunek do równania otrzymujemy: 
\begin{equation*}
v_{x}(0) = v_{0}\cos{\alpha} = C 
\end{equation*}
Otrzymujemy zatem równanie prędkości w czasie w kierunku osi $X$, które podane jest wzorem: 
\begin{equation*}
v_{x}(t) = v_{0}\cos{\alpha} \cdot \mathrm{e}^{-\beta t}
\end{equation*}

Teraz gdy mamy równanie dla prędkości, możemy wyznaczyć równanie dla położenia. W tym celu musimy scałkować nasze równanie

\begin{align*}
v_{x}(t) &= v_{0}\cos{\alpha} \cdot \mathrm{e}^{-\beta t} \\
\frac{\operatorname{d}\!x}{\operatorname{d}\!t} &= v_{0}\cos{\alpha} \cdot \mathrm{e}^{-\beta t} \\
x &= \int v_{0}\cos{\alpha} \cdot \mathrm{e}^{-\beta t} \operatorname{d}\!t \\
x(t) &= v_{0}\cos{\alpha} \left( -\frac{1}{\beta} \mathrm{e}^{-\beta t} + D \right)
\end{align*}

Teraz kolejny raz możemy zastosować warunek początkowy. Piłka w momencie $t=0$ jest na współrzędnej $x=0$. Z tego faktu wyznaczymy stałą $D$

\begin{align*}
x(0) &= 0 = v_{0}\cos{\alpha} \left( -\frac{1}{\beta} + D \right) \\
0 &= -\frac{v_{0}\cos{\alpha}}{\beta} + D\cdot v_{0}\cos{\alpha}  \\
D\cdot v_{0}\cos{\alpha} &= \frac{v_{0}\cos{\alpha}}{\beta} \\
D &= \frac {1}{\beta} 
\end{align*}

Wstawiając $D$ do naszego równania otrzymamy:

\begin{align*}
x &= v_{0}\cos{\alpha} \left( -\frac{1}{\beta} \mathrm{e}^{-\beta t} + \frac{1}{\beta} \right) \\

x &= v_{0}\cos{\alpha} \cdot \frac{1}{\beta} \left( -\mathrm{e}^{-\beta t} + 1 \right)

\end{align*}

Możemy teraz przedstawić kompletne równanie położenia piłki na osi $X$ w czasie, które wygląda tak:
\begin{equation*}
x(t) = \frac{v_{0}\cos{\alpha}}{\beta} \left(1 -\mathrm{e}^{-\beta t}  \right)
\end{equation*}

Teraz tą samą procedurę wykonamy dla osi $Y$. Podstawiając $v_{y}=y'$ otrzymujemy równanie:
\begin{align*}
\frac{\operatorname{d}\!v_{y}}{\operatorname{d}\!t} = -g -\beta \cdot v_{y} \\
\frac{\operatorname{d}\!v_{y}}{\operatorname{d}\!t} +\beta \cdot v_{y}= -g 
\end{align*}
Rozwiążemy to równanie używając czynnika całkującego. Bedzie on wynosić:
\begin{equation*}
\mu=\mathrm{e}^{ \int\beta \operatorname{d}\!t}=\mathrm{e}^{ \beta t}
\end{equation*}
Mnożąc wszystkie składniki równania przez $\mu$ dostajemy:
\begin{align*}
\frac{\operatorname{d}\!v_{y}}{\operatorname{d}\!t} \cdot \mathrm{e}^{ \beta t} + \beta \cdot v_{y} \cdot \mathrm{e}^{ \beta t} = -g \cdot \mathrm{e}^{ \beta t} \\
v_{y}' \cdot \mathrm{e}^{ \beta t} +\cdot v_{y} \cdot \beta\mathrm{e}^{ \beta t} = -g \cdot \mathrm{e}^{ \beta t} \\
\left( v_{y} \cdot \mathrm{e}^{ \beta t}\right)' = -g \cdot \mathrm{e}^{ \beta t}
\end{align*}
Teraz całkując obie strony równania otrzymamy równanie dla prędkości w czasie na osi $Y$.

\begin{align*}
v_{y} \cdot \mathrm{e}^{ \beta t} =\int-g \mathrm{e}^{ \beta t} \operatorname{d}\!t \\
v_{y} \cdot \mathrm{e}^{ \beta t} = -\frac{g}{\beta} \mathrm{e}^{ \beta t} + \ln{E} \\ 
v_{y} = -\frac{g}{\beta} + E\mathrm{e}^{-\beta t} \\
\end{align*}
Możemy teraz podstawić warunek początkowy, którym jest $v_{y}(0)=v_{y_0}=v_{0}\sin{\alpha} $. W ten sposób wyznaczymy stałą $E$ i otrzymamy pełne równanie prędkości

\begin{align*}
v_{y}(0) &= v_{0}\sin{\alpha}= -\frac{g}{\beta} + E \\
E &= v_{0}\sin{\alpha} + \frac{g}{\beta}
\end{align*}
Wstawmy teraz otrzymany wynik do równania 
\begin{align*}
v_{y} = -\frac{g}{\beta} + \left(v_{0}\sin{\alpha} + \frac{g}{\beta}\right) \mathrm{e}^{-\beta t} \\
v_{y} = -\frac{g}{\beta} + v_{0}\sin{\alpha} \cdot \mathrm{e}^{-\beta t} + \frac{g}{\beta} \mathrm{e}^{-\beta t} \\
v_{y}(t) = \mathrm{e}^{-\beta t} \left( v_{0}\sin{\alpha} + \frac{g}{\beta}  \right) - \frac{g}{\beta}
\end{align*}

Mamy już równanie dla prędkości wdłuż osi $Y$, a więc teraz obliczymy równanie dla położenia piłki na osi $Y$ w czasie. Dokonamy tego obustronnie całkując powyższe równanie
\begin{align*}
y = \int \mathrm{e}^{-\beta t} \left( v_{0}\sin{\alpha} + \frac{g}{\beta}  \right) - \frac{g}{\beta} \operatorname{d}\!t \\
y = \int \mathrm{e}^{-\beta t} \cdot v_{0}\sin{\alpha} \operatorname{d}\!t + \int \mathrm{e}^{-\beta t} \cdot \frac{g}{\beta}  \operatorname{d}\!t - \int \frac{g}{\beta} \operatorname{d}\!t \\
y= -\frac{v_{0}\sin{\alpha}}{\beta} \cdot \mathrm{e}^{-\beta t} - \frac{g}{\beta^2} \cdot \mathrm{e}^{-\beta t} - \frac{g}{\beta} \cdot t + F \\
y= -\frac{1}{\beta} \left( v_{0}\sin{\alpha} \cdot \mathrm{e}^{-\beta t} + \frac{g}{\beta} \cdot \mathrm{e}^{-\beta t} + gt \right) + F \\
y= -\frac{1}{\beta} \left(  \mathrm{e}^{-\beta t} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) - gt \right) + F
\end{align*}
Teraz podstawiamy warunek początkowy $y(0)=0$, aby wyznaczyć stałą $F$ i zapisać kompletne równanie położenia dla osi $Y$

\begin{align*}
y(0) = 0 &= -\frac{1}{\beta}   \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right)  + F \\
F &= \frac{1}{\beta} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) \\
\end{align*}
Teraz wstawimy nasz wynik do równania i otrzymamy kompletne równanie położenia na osi $Y$ w czasie
\begin{align*}
y &= -\frac{1}{\beta} \left(  \mathrm{e}^{-\beta t} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) + gt \right) + \frac{1}{\beta} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) \\
y &= -\frac{1}{\beta} \left(  \mathrm{e}^{-\beta t} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) + gt - v_{0}\sin{\alpha} -\frac{g}{\beta} \right) \\
y &= -\frac{1}{\beta} \left(  \mathrm{e}^{-\beta t} \cdot v_{0}\sin{\alpha} + \mathrm{e}^{-\beta t} \cdot \frac{g}{\beta} + gt - v_{0}\sin{\alpha} -\frac{g}{\beta} \right) \\
y &= -\frac{1}{\beta} \left( v_{0}\sin{\alpha} \left( \mathrm{e}^{-\beta t} -1 \right) + \frac{g}{\beta} \left( \mathrm{e}^{-\beta t} -1 \right) +gt \right) \\
y &= -\frac{1}{\beta} \left(\mathrm{e}^{-\beta t} -1 \right) \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) -\frac{gt}{\beta} \\
y &= \frac{1}{\beta} \left(1- \mathrm{e}^{-\beta t}  \right) \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) -\frac{gt}{\beta} \\
\end{align*}
Ostatecznie możemy zapisać nasze równanie położenia na osi $Y$ w czasie jako:
\begin{equation*}
y(t) = \left(1- \mathrm{e}^{-\beta t}  \right) \left( \frac{v_{0}\sin{\alpha}}{\beta} + \frac{g}{\beta^2} \right) -\frac{gt}{\beta} 
\end{equation*}


Wypiszmy sobie nasze równania dla współrzędnych w czasie: 
\begin {align*}
x(t) = \frac{v_{0}\cos{\alpha}}{\beta} \left(1 -\mathrm{e}^{-\beta t}  \right) &&\quad y(t) = \left(1- \mathrm{e}^{-\beta t}  \right) \left( \frac{v_{0}\sin{\alpha}}{\beta} + \frac{g}{\beta^2} \right) -\frac{gt}{\beta} 
\end{align*}
Będziemy teraz chcieli wyznaczyć tor lotu naszej piłki. W tym celu potrzebujemy znaleźć funkcję $y(x)$. Musimy opisać $t$ w równaniu $y(t)$ za pomocą $x$

Składnik $ \left(1 -\mathrm{e}^{-\beta t}  \right) $ wyznaczymy łatwo poprzez podzielenie obu ston równania przez $\frac{v_{0}\cos{\alpha}}{\beta} $.
\begin{align*}
x &= \frac{v_{0}\cos{\alpha}}{\beta} \left(1 -\mathrm{e}^{-\beta t}  \right) \\
\left(1 -\mathrm{e}^{-\beta t}  \right) &= \frac{x\beta}{v_{0}\cos{\alpha}}
\end{align*}

Teraz wyznaczymy $t(x)$ z powyższego równania 
\begin{align*}
\left(1 -\mathrm{e}^{-\beta t}  \right) &= \frac{x\beta}{v_{0}\cos{\alpha}} \\
\mathrm{e}^{-\beta t} &= 1-  \frac{x\beta}{v_{0}\cos{\alpha}} \\
\mathrm{e}^{\beta t} &= \frac{1}{1-\frac{x\beta}{v_{0}\cos{\alpha}}} = \frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \\
\beta t &= \ln{\left \lvert\frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \right \rvert} \\
t &= \frac{1}{\beta} \ln{\left \lvert\frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \right \rvert}
\end{align*}

Możemy teraz wstawić $t$ do równania $y(t)$: 
\begin{align*}
y &= \left(1- \mathrm{e}^{-\beta t}  \right) \left( \frac{v_{0}\sin{\alpha}}{\beta} + \frac{g}{\beta^2} \right) -\frac{gt}{\beta} \\
y &= \frac{x\beta}{v_{0}\cos{\alpha}} \left( \frac{v_{0}\sin{\alpha}}{\beta} + \frac{g}{\beta^2} \right) -\frac{g}{\beta^2} \cdot  \ln{\left \lvert\frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \right \rvert}
\end{align*}
Ostatecznie otrzymujemy zatem równanie toru lotu piłki $y(x)$, które ma postać:
\begin{equation*}
y(x)= \frac{x}{v_{0}\cos{\alpha}} \left( v_{0}\sin{\alpha} + \frac{g}{\beta} \right) -\frac{g}{\beta^2} \cdot  \ln{\left \lvert\frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \right \rvert}
\end{equation*}
Możemy je również zapisać jako: 
\begin{equation*}
y(x)= x{\tg{\alpha}} +  \frac{xg}{\beta v_{0}\cos{\alpha}}  -\frac{g}{\beta^2} \cdot  \ln{\left \lvert\frac{v_{0}\cos{\alpha}}{v_{0}\cos{\alpha}-x\beta} \right \rvert}
\end{equation*}


## $\text{3.1 Metoda Eulera}$

Jest to jedna z najprostszych i najbardziej intuicyjnych metod obliczeniowych, która pomimo swojej podstawowej natury, jest niezwykle skuteczna w sytuacjach, gdzie wymagana jest szybkość i przejrzystość obliczeń, a nie ekstremalna precyzja. W naszym modelu, metoda Eulera stosowana jest do iteracyjnego obliczania kolejnych pozycji piłki na płaszczyźnie 2D, co pozwala na zrozumienie podstawowej dynamiki jej ruchu.<br>Każdy krok w symulacji metody Eulera rozpoczyna się od wywołania funkcji derivative, która oblicza prędkości $v_{x_i}$   i  $v_{y_i}$  oraz przyspieszenia w oparciu o aktualne warunki ruchu piłki, takie jak jej położenie i siły działające na nią (przyspieszenie grawitacyjne i opór powietrza).<br>Następnie, wykorzystując te wartości, metoda Eulera aktualizuje położenie piłki zgodnie z równaniami:
$$
\left\{ \begin{array}{ll}
x_{i+1}=x_{i} + h \cdot v_{x_i}\\
y_{i+1}=y_{i} + h \cdot v_{y_i}\\
\end{array}, \right.
$$
gdzie $h$ to krok czasowy symulacji, wybrany tak, aby zapewnić odpowiedni balans między dokładnością a wydajnością obliczeniową.<br>Dzięki temu, jesteśmy w stanie śledzić trajektorię piłki od momentu kopnięcia, poprzez jej lot w powietrzu, aż do jej potencjalnego lądowania na murawie boiska lub trafienia do bramki.

#### Implementacja metody Eulera prezentuje się w następujący sposób:

```python
# Przekształcenie równań różniczkowych na układ równań pierwszego rzędu
def pochodne(state, t):
    x, y, vx, vy = state
    dvx = -beta * vx
    dvy = -g - beta * vy
    return [vx, vy, dvx, dvy]

# Implementacja metody Eulera z warunkami ograniczającymi
def metoda_eulera(f, w0, t0, tn, h):
    czas = [t0]
    wyniki = [w0]
    t = t0
    while t < tn:
        dw = f(w0, t)
        w0 = [y + h * dwi for y, dwi in zip(w0, dw)]
        t += h
        czas.append(t)
        wyniki.append(w0)
        if w0[1] < 0 or w0[0] >= 11:
            break
    return np.array(wyniki)
```

### Do zinterpretowania wyników potrzebujemy jeszcze odległość bramki od punktu rzutu karnego, czyli $11$ metrów, oraz wysokość bramki, czyli $2,44$ metra. [[3]](#bibliografia3) 
Sprawdzimy jakie trajektorie obierze piłka, dla kątów $ 10^\circ , 15^\circ , 20^\circ, 35^\circ$ oraz dla prędkości początkowych: $20\frac{m}{s} , 25\frac{m}{s} , 30\frac{m}{s}, 35\frac{m}{s}$

```python
goal_height = 2.44  
goal_distance = 11  
angles = [10, 15, 20, 35]  
velocities = [20, 25, 30, 35]  
dt = 0.01

# Przeprowadzenie symulacji dla różnych kątów i prędkości początkowych
for angle in angles:
    plt.figure(figsize=(10, 5))
    plt.title(f"Trajektorie piłki nożnej dla kąta {angle}°")
    plt.xlabel('Dystans [m]')
    plt.ylabel('Wysokość [m]')
    plt.axhline(0, color='green', linestyle='-')  
    plt.plot([goal_distance, goal_distance], [0, goal_height], color='black', linestyle='-',linewidth=2, label='Bramka') 

    for velocity in velocities:
        angle_rad = np.radians(angle)
        initial_state = [0, 0, velocity * np.cos(angle_rad), velocity * np.sin(angle_rad)]
        wyniki = metoda_eulera(pochodne, initial_state, 0, 2, dt)  # Załóżmy, że tn = 2 sekundy
        # Konwersja wyników na format przydatny do wizualizacji
        wyniki_x = [state[0] for state in wyniki]
        wyniki_y = [state[1] for state in wyniki]
        plt.plot(wyniki_x, wyniki_y, label=f'v0={velocity} m/s')
    plt.legend()
    plt.show()
```

## Przykładowa animacja [[4]](#bibliografia4)

```python
# Funkcja do tworzenia i wyświetlania animacji dla danej prędkości i kąta
def animacja(angle, velocity):
    angle_rad = np.radians(angle)
    initial_state = [0, 0, velocity * np.cos(angle_rad), velocity * np.sin(angle_rad)]
    trajectory = metoda_eulera(pochodne, initial_state, 0, 2, dt)
 
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_title(f"Trajektoria piłki dla kąta {angle}° i prędkości {velocity} m/s")
    ax.set_xlabel('Dystans [m]')
    ax.set_ylabel('Wysokość [m]')
    ax.axhline(0, color='green', linestyle='-')
    ax.plot([goal_distance, goal_distance], [0, goal_height], color='black', linestyle='-', linewidth=2)
    line, = ax.plot([], [], '-', lw=2)
    ball, = ax.plot([], [], 'o', markersize=15, markerfacecolor='white', markeredgecolor='black')
    ax.set_xlim(0, 11.5)

 
    def init():
        line.set_data([], [])
        ball.set_data([], [])
        return line, ball
 
    def update(frame):
        xdata = trajectory[:frame, 0]
        ydata = trajectory[:frame, 1]
        line.set_data(xdata, ydata)
        ball.set_data(trajectory[frame, 0], trajectory[frame, 1])
        return line, ball
 
    ani = FuncAnimation(fig, update, frames=len(trajectory), init_func=init, blit=True, interval=50)
    return ani
 
angle = 15
velocity= 30
ani = animacja(angle, velocity)
HTML(ani.to_jshtml())
```

## $ \text{3.2 Analiza błędu metody numerycznej} $

#### Do analizowania błędu metody numerycznej musimy policzyć rozwiązanie numeryczne dla coraz mniejszych kroków czasowych, a następnie policzyć dla każdego maksymalny błąd. Błąd ten będzie liczony względem wyników z metody analitycznej, poprzez równania policzone w pierwszej części projektu. <br> Najpierw napiszemy funkcję, która poda wyniki równania rozwiązanego analitycznie, w konkretnym czasie.

```python
def rozwiazanie_analityczne(t, v0, alpha, g, beta):
    x = (v0 * np.cos(alpha) / beta) * (1 - np.exp(-beta * t))
    y = (1 - np.exp(-beta * t)) * ((v0 * np.sin(alpha) / beta) + (g / beta**2)) - (g * t / beta)
    return [x, y]
```

#### Teraz zmodyfikujemy naszą metodę Eulera, aby nie przerywała się, gdy wyniki dla $x$ przekroczą $11$ i wyniki dla $y$ spadną poniżej $0$, ponieważ nie interesuje nas teraz realizm sytuacji, lecz jedynie policzenie błędu względnego i ustalenie rzędu tego błędu.

```python
def metoda_eulera_do_analizy(f, w0, t0, tn, h):
    czas = [t0]
    wyniki = [w0]
    t = t0
    while t < tn:
        dw = f(w0, t)
        w0 = [y + h * dwi for y, dwi in zip(w0, dw)]
        t += h
        czas.append(t)
        wyniki.append(w0)
    return np.array(wyniki)
```

#### Przyjmujmy warunki początkowe $v_{0}=20\frac{m}{s}$ oraz kąt $ \alpha=20^\circ $ i zaczynamy analizę błędu.

```python
alpha = np.radians(20)
v0 = 20
initial_state = [0, 0, v0 * np.cos(alpha), v0 * np.sin(alpha)]

wyniki_analityczne = []
time_jumps = []
jump = 1
while jump > 0:
    time_jumps.append(jump)
    jump -= 0.01

max_errors = []
for time_jump in time_jumps:
    errors = []
    wyniki_analityczne = []
    t = 0
    max_iterations = int(10 / time_jump) + 1  # Określenie maksymalnej liczby iteracji

    while t <= 10:
        wyniki_analityczne.append(rozwiazanie_analityczne(t, v0, alpha, g, beta))
        t += time_jump
        # Sprawdzanie, czy osiągnięto maksymalną liczbę iteracji
        if len(wyniki_analityczne) >= max_iterations:
            break

    # Dostosowanie długości listy wyników metody Eulera
    wyniki_eulera = metoda_eulera_do_analizy(pochodne, initial_state, 0, 10, time_jump)
    wyniki_eulera = wyniki_eulera[:len(wyniki_analityczne)]

    for i, wynik_analityczny in enumerate(wyniki_analityczne):
        wynik_eulera = wyniki_eulera[i]
        error_x = np.abs(wynik_analityczny[0] - wynik_eulera[0])
        error_y = np.abs(wynik_analityczny[1] - wynik_eulera[1])
        if error_x >= error_y:
            errors.append(error_x)
        else:
            errors.append(error_y)
    max_error = max(errors)
    max_errors.append(max_error)

plt.loglog(time_jumps, max_errors, linestyle='-', label='Błędy numeryczne')
plt.xlabel('Krok czasowy')
plt.ylabel('Błąd')

# Wyznaczenie tangensa kąta nachylenia
coefficients = np.polyfit(np.log(time_jumps), np.log(max_errors), 1)
tangens = coefficients[0]
plt.plot(time_jumps, np.exp(coefficients[1]) * time_jumps**coefficients[0], '--', label='Dopasowana prosta')
plt.title('Zależność błędu od kroku czasowego')
plt.legend()
plt.show()
print(f'Tangens kąta nachylenia: {tangens}')

```

#### Tangens nachylenia wykresu w skali logarytmicznej wynosi około $1$ - tyle właśnie wynosi rząd błędu naszej metody.
Oznacza to, że błąd numeryczny metody Eulera zmienia się liniowo w stosunku do logarytmu kroku czasowego. W miarę zmniejszania kroku czasowego, błąd numeryczny maleje, ale robi to w sposób proporcjonalny do logarytmu zmniejszenia kroku.

## $ \text{3.3 Metoda Rungego-Kutty czwartego rzędu (RK4)} $

### Zajmiemy się teraz analizą błędów numerycznych, porównując wyniki uzyskane za pomocą metody Eulera z wynikami uzyskanymi przy użyciu bardziej zaawansowanej metody numerycznej, takiej jak metoda Rungego-Kutty czwartego rzędu (RK4).

Ta metoda numeryczna jest znana z wyjątkowej dokładności w rozwiązywaniu równań różniczkowych, szczególnie w systemach dynamicznych, gdzie zmiany stanu są znaczne lub nieregularne. RK4 osiąga to przez wykorzystanie czterech różnych estymat pochodnych w obrębie pojedynczego kroku czasowego, aby lepiej przewidzieć przyszły stan systemu. <br> W każdej iteracji metody RK4 obliczane są cztery wartości pochodnych $k_{1}, k_{2}, k_{3}, k_{4}$ które następnie są łączone w finalny krok aktualizacji stanu:
$$
\left\{ \begin{array}{ll}
k_{1}=f(t,w)\\
k_{2}=f(t+\frac{h}{2}, w+\frac{h}{2}k_{1})\\
k_{3}=f(t+\frac{h}{2}, w+\frac{h}{2}k_{2})\\
k_{4}=f(t+h, w+\frac{h}{2}k_{3})\\
\end{array}, \right.
$$
Następnie, pozycja i prędkość piłki są aktualizowane przy użyciu średniej ważonej tych pochodnych:
\begin{equation*}
w_{i+1}=w{i}+\frac{h}{6}(k_{1}+2k_{2}+2k_{3}+k_{4})
\end{equation*}
Dzięki temu skomplikowanemu podejściu, RK4 jest w stanie dostarczyć wysokiej jakości wyniki w symulacjach, co jest niezwykle ważne w analizie, gdzie precyzyjne modelowanie trajektorii może mieć bezpośredni wpływ na ostateczny wynik rzutu karnego. 


#### Pierwszym krokiem będzie implementacja metody RK4. [[5]](#bibliografia5) Następnie przeprowadzimy symulacje dla tych samych warunków początkowych przy użyciu obu metod i porównamy wyniki. Zbadamy również, jak wielkość kroku czasowego (dt) wpływa na dokładność obu metod.


```python
def rk4(f, w0, t0, tn, h):
    czas = [t0]
    wyniki = [w0]
    t = t0
    while t < tn:
        k1 = h * np.array(f(w0, t))
        k2 = h * np.array(f(w0 + 0.5 * k1, t + 0.5 * h))
        k3 = h * np.array(f(w0 + 0.5 * k2, t + 0.5 * h))
        k4 = h * np.array(f(w0 + k3, t + h))
        w0 = w0 + (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h
        czas.append(t)
        wyniki.append(w0)
        if w0[1] < 0 or w0[0] >= goal_distance:
            break
    return np.array(wyniki)
```

```python
initial_velocity = 30  
initial_angle = 15  
angle_rad = np.radians(initial_angle)
initial_state = [0, 0, initial_velocity * np.cos(angle_rad), initial_velocity * np.sin(angle_rad)]

# Symulacja przy użyciu metody Eulera i RK4
dt = 0.01 
trajectory_euler = metoda_eulera(pochodne, initial_state, 0, 5, dt)
trajectory_rk4 = rk4(pochodne, initial_state, 0, 5, dt)

plt.figure(figsize=(10, 5))
plt.plot(trajectory_euler[:, 0], trajectory_euler[:, 1], label='Metoda Eulera', linestyle='--')
plt.plot(trajectory_rk4[:, 0], trajectory_rk4[:, 1], label='Metoda RK4', linestyle='-')
plt.axhline(0, color='green', linestyle='-')
plt.plot([goal_distance, goal_distance], [0, goal_height], color='black', linestyle='-', linewidth=2, label='Bramka')
plt.title(f"Porównanie trajektorii (v0={initial_velocity} m/s, kąt={initial_angle}°)")
plt.xlabel('Dystans [m]')
plt.ylabel('Wysokość [m]')
plt.legend()
plt.show()
```

```python
dt_values = [0.01, 0.05, 0.1]
angle_values = [20, 30, 45]
velocity_values = [20, 35, 50]
results = {}

for dt in dt_values:
    for angle in angle_values:
        for velocity in velocity_values:
            angle_rad = np.radians(angle)
            initial_state = [0, 0, velocity * np.cos(angle_rad), velocity * np.sin(angle_rad)]
            trajectory_euler = metoda_eulera(pochodne, initial_state, 0, 5, dt)
            trajectory_rk4 = rk4(pochodne, initial_state, 0, 5, dt)
            results[(dt, angle, velocity)] = {'Euler': trajectory_euler, 'RK4': trajectory_rk4}

# Wyświetlenie wyników dla przykładowego zestawu parametrów
first_key = list(results.keys())[0]
example_trajectory_euler = results[first_key]['Euler']
example_trajectory_rk4 = results[first_key]['RK4']

selected_keys = list(results.keys())[::3]  # Wybieramy co trzeci zestaw parametrów do wizualizacji

for key in selected_keys:
    dt, angle, velocity = key
    trajectory_euler = results[key]['Euler']
    trajectory_rk4 = results[key]['RK4']

    plt.figure(figsize=(10, 5))
    plt.plot(trajectory_euler[:, 0], trajectory_euler[:, 1], label='Metoda Eulera', linestyle='--')
    plt.plot(trajectory_rk4[:, 0], trajectory_rk4[:, 1], label='Metoda RK4', linestyle='-')
    plt.axhline(0, color='green', linestyle='-')
    plt.plot([goal_distance, goal_distance], [0, goal_height], color='black', linestyle='-', linewidth=2, label='Bramka')
    plt.title(f"Porównanie trajektorii (dt={dt}, kąt={angle}°, v0={velocity} m/s)")
    plt.xlabel('Dystans [m]')
    plt.ylabel('Wysokość [m]')
    plt.legend()
    plt.show()
```

Metoda Eulera i metoda Rungego-Kutty czwartego rzędu (RK4) generalnie dają zbliżone wyniki dla trajektorii lotu piłki w badanych warunkach. To sugeruje, że dla małych kroków czasowych i stosunkowo prostych problemów fizycznych, metoda Eulera może być wystarczająco dokładna.
<br>Metoda RK4, będąca metodą wyższego rzędu, zazwyczaj oferuje większą dokładność i stabilność, szczególnie przy większych krokach czasowych lub bardziej złożonych problemach dynamicznych.
Zmniejszanie kroku czasowego (dt) zwiększa dokładność obu metod, ale także zwiększa czas obliczeń. Znalezienie optymalnego kroku czasowego, który zapewnia akceptowalny kompromis między dokładnością a wydajnością obliczeniową, jest kluczowe.

### Uderzając piłkę z rzutu karnego, istnieją dwie najlepsze strategie: albo uderzamy piłkę w bok bramki i idealnie pod poprzeczkę, albo uderzamy ją tak, aby wpadła przy ziemi koło słupka. W naszej symulacji nie możemy jednak sprawdzić strategii, w której piłka leci w bok, więc zbadamy trajektorię piłki, która kończy swój lot tuż nad linią bramkową.
```python
velocity = 30  # prędkość początkowa w m/s
angle_range = np.arange(1, 90, 0.1)  # zakres kątów od 1° do 90° z krokiem co 0.1°

# Szukanie odpowiedniego kąta
for angle in angle_range:
    angle_rad = np.radians(angle)
    initial_state = [0, 0, velocity * np.cos(angle_rad), velocity * np.sin(angle_rad)]
    
    # Symulacja trajektorii
    trajektoria = metoda_eulera(pochodne, initial_state, 0, 2, 0.01)
    x, y = trajektoria[-1, 0], trajektoria[-1, 1]
    
    # Sprawdzenie, czy piłka trafiła do bramki
    if 0.05 < y < goal_height and x >= goal_distance:
        optimal_angle = round(angle,2)
        break

plt.figure(figsize=(10, 5))
plt.plot(trajektoria[:, 0], trajektoria[:, 1], label=f"Trajektoria piłki")
plt.axhline(0, color='green', linestyle='-') 
plt.plot([goal_distance, goal_distance], [0, goal_height], color='black', linestyle='-', linewidth=2, label='Bramka')
plt.plot(x, y, 'o', markersize=15, markerfacecolor='white', markeredgecolor='black')
plt.title(f"Trajektoria piłki przy prędkości {velocity} m/s i kącie {optimal_angle:.1f}°")
plt.xlabel('Dystans [m]')
plt.ylabel('Wysokość [m]')
plt.legend()
plt.show()

print(f"Optymalny kąt strzału: {optimal_angle} stopni")
```

## $ \text{6. Problem w praktyce - znalezienie zakresu kątów, dla których trafimy do bramki z "rekordową" prędkością początkową} $

Wykonawcy rzutów karnych potrafią nadać piłce prędkości dochodzące do $80 mph$ [[6]](#bibliografia6) , co jest równe $35.76 \frac{m}{s}$.<br>Dla tej prędkości znajdziemy zakres kątów, dla których piłka trafi do bramki.

```python
def znajdz_odpowiedni_kat(velocity, goal_distance, goal_height):
    angle_range = np.arange(1, 90, 0.1)  # zakres kątów od 1° do 90° z krokiem co 0.1°
    valid_angles = []

    for angle in angle_range:
        angle_rad = np.radians(angle)
        initial_state = [0, 0, velocity * np.cos(angle_rad), velocity * np.sin(angle_rad)]
        trajectory = metoda_eulera(pochodne, initial_state, 0, 2, 0.01)
        
        # Przechodzimy przez trajektorię i sprawdzamy warunki trafienia w bramkę
        for x, y in zip(trajectory[:, 0], trajectory[:, 1]):
            if x >= goal_distance and 0 <= y < goal_height:
                valid_angles.append(round(angle,2))
                break  # Zatrzymujemy symulację, gdy piłka wpadnie do bramki

    return valid_angles

fastest_kick_speed_m_s = 35.76  # prędkość w m/s

# Znajdź zakres kątów, dla których piłka wpadnie do bramki
valid_angles = znajdz_odpowiedni_kat(fastest_kick_speed_m_s, goal_distance, goal_height)
print(valid_angles)
```
